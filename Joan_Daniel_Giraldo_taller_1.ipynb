## Parte 2: Análisis con Numpy  

11. Convierte las columnas de estadísticas (`HP`, `Attack`, `Defense`, `Sp. Atk`, `Sp. Def`, `Speed`) en un array de Numpy.  

````python
import numpy as np

stats_pokemon= df[["HP","Attack","Defense","Sp. Atk","Sp. Def","Speed"]].to_numpy()

stats_pokemon
````

12. Calcula la media, mediana y desviación estándar de cada estadística usando Numpy.  

````python
media= np.mean(stats_pokemon, axis=0)
mediana=np.median(stats_pokemon, axis=0)
desviacion=np.std(stats_pokemon, axis=0)

print("media:",media)
print("mediana;", mediana)
print("desviacion:", desviacion)
````

13. Normaliza las estadísticas dividiendo cada valor por el máximo de su columna.

````python
maximos= np.max(stats_pokemon, axis= 0)

normal= stats_pokemon/maximos

normal
````

14. Encuentra los índices de los 5 Pokémon con mayor defensa (`Defense`).  

````python
Defensa= stats_pokemon[:,2]
indices_ordenados=np.argsort(Defensa)

top_5_indices_mas_duros= indices_ordenados[-5:][::-1]

top_5_indices_mas_duros
````

15. Calcula la correlación entre `Attack` y `Defense`.

````python
Ataque= stats_pokemon[:,1]
Defensa1=stats_pokemon[:,2]

Correlacion= np.corrcoef(Ataque, Defensa1)

Correlacion
````

16. Usa Numpy para encontrar cuántos Pokémon tienen una suma de (`HP`, `Attack` y `Defense`) superior a 300.

````python
Columnas=stats_pokemon[:,0:3]
Suma=np.sum(Columnas, axis=1)

Pokemones_superiores=np.sum(Suma > 300)

Pokemones_superiores
````
